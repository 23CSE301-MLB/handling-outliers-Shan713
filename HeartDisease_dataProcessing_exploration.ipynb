```python
import seaborn as sns
import matplotlib.pyplot as plt

df_deathEvent = df[df['DEATH_EVENT']==1]

sns.histplot(data = df_deathEvent, x='age',bins=20)
plt.title("Age Distribution for occurance of Death")
plt.show()
```

```python
df_deathEvent['category_gender'] = pd.cut(df_deathEvent['sex'],bins=2,labels=['male','female'])
sns.countplot(data= df_deathEvent, x='category_gender')
plt.title("Gender Dist over occurance of Death Event")
plt.xlabel("Gender")
plt.show()
```

```python
sns.scatterplot(data= df, y='ejection_fraction', x='age', hue='DEATH_EVENT')
plt.title("Relation between Age and Ejection Fraction")
plt.show()
```

```python
sns.boxplot(data=df, y='creatinine_phosphokinase')
```

```python
Q1 = df['creatinine_phosphokinase'].quantile(0.25)
Q3 = df['creatinine_phosphokinase'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df['creatinine_phosphokinase'] < lower_bound) | (df['creatinine_phosphokinase'] > upper_bound)]
print(outliers_iqr.shape)
```

```python
import numpy as np
df_capped = df.copy()
df_capped['creatinine_phosphokinase'] = np.where(df_capped['creatinine_phosphokinase'] > upper_bound, upper_bound,
                                    np.where(df_capped['creatinine_phosphokinase'] < lower_bound, lower_bound, df_capped['creatinine_phosphokinase']))

outliers_iqr2 = df_capped[(df_capped['creatinine_phosphokinase'] < lower_bound) | (df_capped['creatinine_phosphokinase'] > upper_bound)]
print(outliers_iqr2.shape)

sns.boxplot(data=df_capped, y='creatinine_phosphokinase')
```

```python
sns.pairplot(df_capped[['age','platelets','DEATH_EVENT']],hue='DEATH_EVENT')
plt.show()
```

```python
df_capped.sort_values('age',inplace=True)
df_capped
df_capped.to_csv('processed_heart_diesease_Dataset.csv',index=False)
```
